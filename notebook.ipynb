## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

***"Faktor apa saja yang memengaruhi penurunan conversion rate pada pengguna aplikasi Android di wilayah Jabodetabek sebesar 5% selama periode Flash Sale Maret 2026?"***

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

### Cleaning Data

#### Fixing problem

```python
# Membersihkan data orders_df

# 1. Mengganti tipe data menjadi datetime()
datetime_column = ["order_purchase_timestamp"]

for column in datetime_column:
    orders_df[column] = pd.to_datetime(orders_df[column])

# 2. Filter data yang hanya berstatus "delivered" dan tahun 2017 & 2018
orders_df = orders_df[
    (orders_df['order_status'] == 'delivered') &
    (orders_df['order_purchase_timestamp'].dt.year.isin([2017, 2018]))
].copy()

# 3. Menghapus kolom data yang tidak diperlukan
orders_df.drop(columns=["order_status", "order_approved_at", "order_delivered_carrier_date", 
                        "order_delivered_customer_date", "order_estimated_delivery_date"], inplace=True)

# Periksa hasil cleaning
orders_df.info()
orders_df.describe(include='all')
```

```python
# Membersihkan dataset items_df
# 1. Menghapus kolom data yang tidak diperlukan
items_df.drop(columns=["seller_id", "shipping_limit_date"], inplace=True)

# Periksa hasil cleaning
items_df.info()      
```

```python
# Membersihkan dataset product_df
# 1. Mengisi missing value pada kategori product
product_df["product_category_name"].fillna(value="others", inplace=True)

# 2. Menghapus kolom data yang tidak diperlukan
product_df.drop(columns=["product_weight_g", "product_length_cm", "product_height_cm", 
                         "product_width_cm", "product_photos_qty", "product_description_lenght", 
                         "product_name_lenght"], inplace=True)  

# Periksa hasil cleaning
product_df.info()
```

```python
# Menggabungkan data product_df dan translation_df
# 1. Melakukan penyesuaian antar nama kolom agar sesuai (semua memakai lowercase dan tetap pakai underscore)
product_df["product_category_name"] = product_df["product_category_name"].str.lower()
translation_df["product_category_name"] = translation_df["product_category_name"].str.lower()

# 2. Menggabungkan data product_df dan translation_df
product_translation_df = pd.merge(
    left=product_df,
    right=translation_df,
    how="left",
    on="product_category_name" #key
)

# 3. Mengisi data terjemahan pada translation_df dengan bahasa asli dari product_df.
product_translation_df["product_category_name_english"].fillna(product_translation_df["product_category_name"])

# 4. Menghapus kolom product_category_name
product_translation_df.drop(columns=["product_category_name"], inplace=True)

# Periksa hasil cleaning
product_translation_df.head()
```

```python
# Menghitung total orders dan revenue per bulan 
# 1. Ekstraksi bulan dan tahun
new_dataset_df["month"] = new_dataset_df["order_purchase_timestamp"].dt.month
new_dataset_df["year"] = new_dataset_df["order_purchase_timestamp"].dt.year

# 2. Filter data khusus tahun 2017
df_2017 = new_dataset_df[new_dataset_df["year"] == 2017]

# 3. Total orders dan revenue per bulan
monthly_sales = df_2017.groupby("month").agg(
    total_orders=("order_id", "nunique"),
    total_revenue=("revenue", "sum")
).reset_index().sort_values("month")

print(monthly_sales)
```

**Insight:** Eksplorasi ini dilakukan untuk melihat tren peforma penjualan dari bulan ke bulan sepanjang 2017. Hal ini dapat mengidentifikasi di bulan mana saja terjadi lonjakan pembelian dan pendapatan.

```python
# Revenue tertinggi dan terendah per kategori tahun 2017
# 1. Ekstraksi bulan dan tahun
new_dataset_df["month"] = new_dataset_df["order_purchase_timestamp"].dt.month
new_dataset_df["year"] = new_dataset_df["order_purchase_timestamp"].dt.year

# 2. Filter data khusus tahun 2017
df_2017 = new_dataset_df[new_dataset_df["year"] == 2017]

# 3. Total penjualan per kategori (semua bulan digabung)
revenue_category = df_2017.groupby("product_category_name_english").agg(
    total_orders=("order_id", "nunique"),
    total_revenue=("revenue", "sum")
).reset_index().sort_values("total_orders", ascending=False)

# 4. 3 kategori tertinggi dan terendah berdasarkan revenue
top_r = revenue_category.nlargest(3, "total_revenue")
bottom_r = revenue_category.nsmallest(3, "total_revenue")

print("========Top 3 Kategori Berdasarkan Revenue Tertinggi=========")
print(top_r)

print("\n======Bottom 3 Kategori Berdasarkan Revenue Terendah=======")
print(bottom_r)
```

**Insight:** Pemetaan 3 kategori dengan pendapatan tertinggi dan terendah digunakan untuk memetakan kategori yang perlu terus ditingkatkan kualitasnya dan yang perlu mendapatkan evaluasi kembali karena kontribusinya terhadap omzet yang sedikit.

```python
# Pesanan tertinggi dan terendah per kategori tahun 2017
# 1. Ekstraksi bulan dan tahun
new_dataset_df["month"] = new_dataset_df["order_purchase_timestamp"].dt.month
new_dataset_df["year"] = new_dataset_df["order_purchase_timestamp"].dt.year

# 2. Filter data khusus tahun 2017
df_2017 = new_dataset_df[new_dataset_df["year"] == 2017]

# 3. Total penjualan per kategori (semua bulan digabung)
orders_category = df_2017.groupby("product_category_name_english").agg(
    total_orders=("order_id", "nunique"),
    total_revenue=("revenue", "sum")
).reset_index().sort_values("total_orders", ascending=False)

# 4. 3 kategori tertinggi dan terendah berdasarkan pesanan
top_o = orders_category.nlargest(3, "total_orders")
bottom_o = orders_category.nsmallest(3, "total_orders")

print("========Top 3 Kategori Pesanan Tertinggi=========")
print(top_o)

print("\n======Bottom 3 Kategori Pesanan Terendah=======")
print(bottom_o)
```

**Insight:** Pemetaan 3 kategori dengan jumlah pesanan dilakukan untuk melihat produk yang paling populer dibeli pelanggan dan yang tidak. Tidak semua jumlah pesanan terbanyak memberikan pendapatan tinggi, karena bergantung pada kategorinya.

## Analisis Lanjutan RFM

**Analisis RFM** merupakan salah satu metode yang digunakan untuk melakukan segmentasi pelanggan berdasarkan tiga parameter, yaitu recency, frequency, dan monetary.

- Recency: Menghitung selisih hari antara tanggal terakhir transaksi di seluruh dataset dengan tanggal terakhir pelanggan X belanja.
- Frequency: Menghitung berapa kali order_id unik muncul untuk setiap pelanggan.
- Monetary: Menjumlahkan total pendapatan kotor (revenue) yang dihasilkan dari pelanggan tersebut.

Pelanggan high value merupakan Pelanggan yang Recency-nya rendah (baru belanja), Frequency-nya tinggi (sering), dan Monetary-nya tinggi (belanja banyak). 
Low Value merupakan pelanggan yang sudah lama tidak belanja (Recency tinggi) dan cuma belanja sedikit sekali.

```python
# Menghitung RFM berdasarkan skoring
# 1. Hapus data dengan acuan data order yang sudah difilter berdasarkan "delivered"
rfm_clean_data = new_dataset_df.dropna(subset=["order_id"]).copy()

# 2. Tentukan tanggal referensi
reference_date = rfm_clean_data["order_purchase_timestamp"].max() + pd.Timedelta(days=1)

# 3. Membuat group data
rfm_df = rfm_clean_data.groupby(by="customer_unique_id", as_index=False).agg({
    "order_purchase_timestamp": lambda x: (reference_date - x.max()).days,
    "order_id": "nunique",
    "revenue": "sum"
})

# 4. Rename kolom
rfm_df.columns = ["customer_id", "recency", "frequency", "monetary"]

# 5. Scoring (r_score, f_score, m_score)
rfm_df["r_score"] = pd.qcut(rfm_df["recency"].rank(method="first"), 5, labels=[5, 4, 3, 2, 1])
rfm_df["f_score"] = pd.qcut(rfm_df["frequency"].rank(method="first"), 5, labels=[1, 2, 3, 4, 5])
rfm_df["m_score"] = pd.qcut(rfm_df["monetary"].rank(method="first"), 5, labels=[1, 2, 3, 4, 5])

# 6. Total Score & Category
rfm_df["rfm_total_score"] = (
    rfm_df["r_score"].astype(int) + 
    rfm_df["f_score"].astype(int) + 
    rfm_df["m_score"].astype(int)
)

# 6. Menentukan high and low value berdasarkan median (agar tidak menghilangkan data yang sangat besar dan kecil)
threshold = rfm_df["rfm_total_score"].median()
rfm_df["customer_category"] = rfm_df["rfm_total_score"].apply(
    lambda x: "High Value Customer" if x >= threshold else "Low Value Customer"
)

rfm_df.head()
```

**Insight:** Analisis profil pelanggan ini dilakukan dengan metode RFM Scoring, di mana setiap pelanggan diberikan peringkat 1-5 berdasarkan Recency (waktu transaksi terakhir), Frequency (jumlah transaksi), dan Monetary (total nilai belanja). Skor dari ketiga variabel tersebut kemudian diakumulasikan untuk mendapatkan skor total yang mencerminkan nilai strategis pelanggan terhadap bisnis. Penentuan kategori didasarkan pada nilai median skor total, di mana pelanggan dengan nilai di atas median diklasifikasikan sebagai High Value Customer, sedangkan sisanya dikategorikan sebagai Low Value Customer.
